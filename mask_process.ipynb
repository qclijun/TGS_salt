```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
%matplotlib inline
import cv2
```

```python
mask_file = "data/train/masks/3e84576386.png"
img = cv2.imread(mask_file, cv2.IMREAD_GRAYSCALE)
kernel = np.ones((3, 3), np.uint8)
dilation = cv2.dilate(img, kernel, iterations=1)
erosion = cv2.erode(img, kernel, iterations=1)

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img, cmap="gray")
axs[0].set_title("original image")
axs[1].imshow(erosion, cmap="gray")
axs[1].set_title("erosion image")
axs[2].imshow(dilation, cmap="gray")
axs[2].set_title("dialation image")
```

```python
def calc_borders(mask, border_sz=3):
    kernel = np.ones((border_sz, border_sz), np.uint8)
    erosion = cv2.erode(mask, kernel, iterations=1)
    return mask - erosion
```

```python
mask_file = "data/train/masks/3e84576386.png"
img = cv2.imread(mask_file, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray")

border = calc_borders(img, 3)
plt.imshow(border, cmap="gray")
```
