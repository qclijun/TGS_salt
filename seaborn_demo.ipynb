```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")
```

```python
sns.__version__
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(x="total_bill", y="tip", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="smoker", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", hue="size", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
df = pd.DataFrame({'time': range(500), 'value': np.random.randn(500).cumsum()})
```

```python
df.head()
```

```python
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", kind="line", data=fmri, dashes=False, markers=True)
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            col="subject", col_wrap=5,
            height=3, aspect=0.75, linewidth=2.5,
            kind="line", data=fmri.query("region=='frontal'"))
```

```python
sns.catplot(x="day", y="total_bill", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", jitter=False, data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

```python
sns.catplot(x="size", y="total_bill", data=tips[tips['size'] != 3])
```

```python
sns.catplot(x="smoker", y="tip", order=["Yes", "No"], data=tips)
```

```python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker", kind="box", data=tips)
```
